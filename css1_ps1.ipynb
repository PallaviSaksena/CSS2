# 3: Remove spaces and change word cases

```
name = " jOhn sERenCes "

# Strip the empty spaces at the start and end of the string
print (name.strip())

# Convert name into title case
print (name.title())

# Convert name into all lowercase
print (name.lower())

# Convert name into all uppercase
print (name.upper())
```
