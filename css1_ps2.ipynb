# 2: Sort, Pop, Add, and Remove things from a List

```
famous_people = ['Zendaya','Olivia', 'Harry']

# Permanently sort list alphabetically

famous_people.sort()
print (famous_people)

# Pop the middle name and store as another variable

singer = famous_people.pop(1)
print (singer)

# Add two new names to the end

famous_people.append('Selena')
famous_people.append('Chris')
print (famous_people)

# Remove the second name

del famous_people[1]
print (famous_people)

# Add your name to 3rd position

famous_people.insert(2, 'Pallavi')
print (famous_people)
```

# 3: Indexing Lists - Even and Odd Numbers

```
integers = list(range(0,21))
print (integers)

# List that indexes only even numbers

print (integers[0:21:2])

# List that indexes only odd numbers

print (integers[1:21:2])

# List that indexes even numbers backwards

print (integers[::-2])

# List that indexes odd numbers backwards

print (integers[-2:-21:-2])
```

# 4: Copy, Delete, and Change Order within Lists

```
# Create a list with numbers from 0 to 35 in increments of 5

list1 = list(range(0, 40, 5))
print (list1)

# Copy the list and delete the last two entries

list1_copy = list1.copy()
del list1_copy[-1]
del list1_copy[-1]

print (list1)
print (list1_copy)

# Create a list with numbers from 0 to -28 in increments of -4

list2 = list(range(0, -29, -4))
print (list2)

# Sort list 2 into increasing order

print (list2[::-1])
```

# 5: Creating Tables with For Loops

```
# Create multiplication table of 17

table_17 = []

factors = list(range(1,11))

for num1 in factors:
  table_17.append(num1*17)

print (table_17)

# Create multiplication table of 19

factors_19 = []
table_19 = []

for num1 in table_17:
  factors_19.append(num1//17)

for num1 in factors_19:
  table_19.append(num1*19)

print (table_19)

# Create table of squares of table_17

tablesq_17 = []

for num1 in table_17:
  tablesq_17.append(num1**2)

print (tablesq_17)
```
