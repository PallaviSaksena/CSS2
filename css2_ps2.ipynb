Block 1: Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Block 2: Creating the Figure

```python
anscombe_df = sns.load_dataset('anscombe')
anscombe_df
```

* different colors for the different datasets
* labels on all the axes
* the same x and y limits for each plot
* a title that has the specific number of the dataset
* text inside the plot that has the mean and standard deviation for both the x and y variables

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(7, 7))
fig.suptitle("Anscombe's Quartet", x = 0.5, y = 1.05)

# for loop to set x and y axis scales

for i in range(0, 1):
  for j in range(0, 1):
    axes[i, j].set_xticks([4, 6, 8, 10, 12, 14, 16, 18])
    axes[i, j].set_yticks([4, 6, 8, 10, 12])

# to make the text inside plots

x = anscombe_df['x']
y = anscombe_df['y']

x_mean = round(np.mean(x), 2)
x_std = round(np.std(x), 2)
y_mean = round(np.mean(y), 2)
y_std = round(np.std(y), 2)

# dataset I

axes[0, 0].plot(x[0:11], y[0:11], linestyle = '', marker = 'o', c= 'red')
axes[0, 0].set_title('Dataset I')
axes[0, 0].set_xlabel('X1')
axes[0, 0].set_ylabel('Y1')

axes[0, 0].text(4, 12, 'x mean:')
axes[0, 0].text(8, 12, x_mean)
axes[0, 0].text(4, 11, 'x std:')
axes[0, 0].text(7, 11, x_std)
axes[0, 0].text(14, 5, 'y mean:')
axes[0, 0].text(18, 5, y_mean)
axes[0, 0].text(14, 4, 'y std:')
axes[0, 0].text(17, 4, y_std)

# dataset II

axes[0, 1].plot(x[11:22], y[11:22], linestyle = '', marker = 'o', c= 'orange')
axes[0, 1].set_title('Dataset II')
axes[0, 1].set_xlabel('X2')
axes[0, 1].set_ylabel('Y2')

axes[0, 1].text(4, 12, 'x mean:')
axes[0, 1].text(8, 12, x_mean)
axes[0, 1].text(4, 11, 'x std:')
axes[0, 1].text(7, 11, x_std)
axes[0, 1].text(13, 5, 'y mean:')
axes[0, 1].text(17, 5, y_mean)
axes[0, 1].text(13, 4, 'y std:')
axes[0, 1].text(17, 4, y_std)

# dataset III

axes[1, 0].plot(x[22:33], y[22:33], linestyle = '', marker = 'o', c= 'blue')
axes[1, 0].set_title('Dataset III')
axes[1, 0].set_xlabel('X3')
axes[1, 0].set_ylabel('Y3')

axes[1, 0].text(4, 11, 'x mean:')
axes[1, 0].text(8, 11, x_mean)
axes[1, 0].text(4, 10, 'x std:')
axes[1, 0].text(7, 10, x_std)
axes[1, 0].text(13, 5, 'y mean:')
axes[1, 0].text(17, 5, y_mean)
axes[1, 0].text(13, 4, 'y std:')
axes[1, 0].text(17, 4, y_std)

# dataset IV
axes[1, 1].plot(x[33:44], y[33:44], linestyle = '', marker = 'o', c= 'purple')
axes[1, 1].set_title('Dataset IV')
axes[1, 1].set_xlabel('X4')
axes[1, 1].set_ylabel('Y4')

axes[1, 1].text(11, 9, 'x mean:')
axes[1, 1].text(15, 9, x_mean)
axes[1, 1].text(11, 8, 'x std:')
axes[1, 1].text(15, 8, x_std)
axes[1, 1].text(11, 6, 'y mean:')
axes[1, 1].text(15, 6, y_mean)
axes[1, 1].text(11, 5, 'y std:')
axes[1, 1].text(15, 5, y_std)

fig.tight_layout()
```

Block 3: Saving the Image
